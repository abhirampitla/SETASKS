Weather Modelling using Waterfall Model


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function to represent the waterfall curve
def quadratic(x, a=-1, b=0, c=10):
    return a * x**2 + b * x + c

# Define the stages of the Waterfall Model
stages = ["Data Collection", "Data Cleaning", "Feature Engineering", "Model Training", "Evaluation"]
x_positions = np.linspace(-2, 2, len(stages))  # Equally spaced x positions
y_positions = quadratic(x_positions)          # Compute y positions using the quadratic function

# Plot the quadratic curve
x_curve = np.linspace(-2.5, 2.5, 500)
y_curve = quadratic(x_curve)
plt.figure(figsize=(12, 6))
plt.plot(x_curve, y_curve, color="blue", linestyle="--", label="Quadratic Curve")

# Add markers and connect them with vertical lines to create the waterfall effect
for i in range(len(stages)):
    if i != 0:
        # Draw a vertical drop to simulate a waterfall
        plt.plot(
            [x_positions[i-1], x_positions[i]],  # Horizontal segment
            [y_positions[i-1], y_positions[i-1]],
            color="blue",
            linewidth=2
        )
        plt.plot(
            [x_positions[i], x_positions[i]],    # Vertical drop
            [y_positions[i-1], y_positions[i]],
            color="blue",
            linewidth=2
        )

# Place markers and labels
plt.scatter(x_positions, y_positions, color="red", zorder=5)
for i, stage in enumerate(stages):
    plt.text(x_positions[i], y_positions[i] + 0.3, stage, ha="center", fontsize=10)

# Style the plot
plt.title("Weather Modeling Representation Using Waterfall Model", fontsize=14)
plt.xlabel("Stages", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend()
plt.show()
```
